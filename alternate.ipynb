```python
import pandas as pd
import os

# data = pd.read_csv(os.path.join("datasets", "medquad.csv"))
data = pd.read_csv('/content/medquad.csv')
```

```python
# Lowercase letters
data["question"] = data["question"].str.lower()
data["answer"] = data["answer"].str.lower()
data["source"] = data["source"].str.lower()
data["focus_area"] = data["focus_area"].str.lower()
```

```python
data
```

```python
# Remove urls (found in answer)
import re

def remove_url(text):
    # Source for finding regex of url pattern of any domain: https://sparkbyexamples.com/pandas/convert-pandas-column-to-lowercase/?expand_article=1
    urlPattern = r"(https:\/\/www\.|http:\/\/www\.|https:\/\/|http:\/\/)?[a-zA-Z0-9]{2,}(\.[a-zA-Z0-9]{2,})(\.[a-zA-Z0-9]{2,})?"
    return re.sub(urlPattern, "URL", str(text))

data["question"] = data["question"].apply(remove_url)
data["answer"] = data["answer"].apply(remove_url)
data["source"] = data["source"].apply(remove_url)
data["focus_area"] = data["focus_area"].apply(remove_url)
```

```python
data
```

```python
# Remove non alphanumeric
def remove_non_alphanumeric(text):
    nonAlpha = r"[^a-zA-Z0-9\s]"
    return re.sub(nonAlpha, "", str(text))
data["question"] = data["question"].apply(remove_non_alphanumeric)
data["answer"] = data["answer"].apply(remove_non_alphanumeric)
data["source"] = data["source"].apply(remove_non_alphanumeric)
data["focus_area"] = data["focus_area"].apply(remove_non_alphanumeric)
```

```python
data
```

```python
# reduce consecutive letters
def reduce_consecutive_letters(text):
    # Regular expression pattern to match three or more consecutive letters
    sequence_pattern = r'([a-zA-Z])\1{2,}'
    # Replacement pattern to replace consecutive letters with just two letters
    seq_replace_pattern = r'\1\1'
    # Apply the substitution using re.sub
    return re.sub(sequence_pattern, seq_replace_pattern, str(text))

data["question"] = data["question"].apply(reduce_consecutive_letters)
data["answer"] = data["answer"].apply(reduce_consecutive_letters)
data["source"] = data["source"].apply(reduce_consecutive_letters)
data["focus_area"] = data["focus_area"].apply(reduce_consecutive_letters)
```

```python
from nltk import download
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

download("punkt")
download("stopwords")
download("wordnet")
```

```python
def remove_stopwords(text):
    """ Stop words removal function """
    text_without_stopwords = ""
    textLst = text.split()
    stopwords_list = stopwords.words("english")
    for word in textLst:
        if word not in stopwords_list:
            text_without_stopwords += word + " "
    return text_without_stopwords

data["question"] = data["question"].apply(remove_stopwords)
data["answer"] = data["answer"].apply(remove_stopwords)
data["source"] = data["source"].apply(remove_stopwords)
data["focus_area"] = data["focus_area"].apply(remove_stopwords)
```

```python
data
```

```python
def lemmatizeText(text):
    """ Lemmatize words """
    lemmatizer = WordNetLemmatizer()
    return lemmatizer.lemmatize(text)


data["question"] = data["question"].apply(lemmatizeText)
data["answer"] = data["answer"].apply(lemmatizeText)
data["source"] = data["source"].apply(lemmatizeText)
data["focus_area"] = data["focus_area"].apply(lemmatizeText)
```

```python
from sklearn.model_selection import train_test_split
train_df, val_df = train_test_split(data, test_size=0.1, random_state=42)
```

```python
# Stemming
import numpy as np
ps = PorterStemmer()
```

```python
data_dict = dict()
#{k: [{k1:[v1, v2, v3]}, {k2:[v1, v2, v3]}]}
corpus = []

for disease_index in range(len(data.focus_area)):
    # Case where disease not in dict
    # print(data.focus_area[disease_index])
    tokenized_answer = word_tokenize(data.answer[disease_index])
    stemmed_answer = []
    for word in tokenized_answer:
        stemmed_answer.append(ps.stem(word))
    if data.focus_area[disease_index] not in data_dict:
        data_dict[data.focus_area[disease_index]] = [{data.question[disease_index]: data.answer[disease_index]}]
    else:
        data_dict[data.focus_area[disease_index]].append({data.question[disease_index]: data.answer[disease_index]}) # Assume no duplicate questions (repetition)\
    corpus.extend(stemmed_answer)
```

```python
train_corpus = []
train_data_dict = dict()

for disease_index in range(len(train_df.focus_area)):
  tokenized_answer = word_tokenize(data.answer[disease_index])
  stemmed_answer = []
  for word in tokenized_answer:
      stemmed_answer.append(ps.stem(word))
  if data.focus_area[disease_index] not in train_data_dict:
      train_data_dict[data.focus_area[disease_index]] = [{data.question[disease_index]: data.answer[disease_index]}]
  else:
      train_data_dict[data.focus_area[disease_index]].append({data.question[disease_index]: data.answer[disease_index]})
  train_corpus.extend(stemmed_answer)
```

```python
# Encode the training and validation data using the tokenizer
train_encodings = tokenizer(train_corpus, truncation=True, padding=True)
val_encodings = tokenizer(val_corpus, truncation=True, padding=True)
```

```python
import torch

class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = CustomDataset(train_encodings, train_df)
val_dataset = CustomDataset(val_encodings, val_df)
```

```python
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',          # output directory
    num_train_epochs=3,              # total number of training epochs
    per_device_train_batch_size=16,  # batch size per device during training
    per_device_eval_batch_size=64,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)

model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = Trainer(
    model=model,                         # the instantiated Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=val_dataset             # evaluation dataset
)

trainer.train()
```
